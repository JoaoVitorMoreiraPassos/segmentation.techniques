```python
import pandas as pd

from glob import glob
from skimage.io import imread
from skimage.measure import label, regionprops_table
```

```python
def extract_props_from_image(mask):
    
    label_image = label(mask)
    
    feature_names = ['bbox_area',
                     'solidity',
                     'centroid', 
                     'orientation',
                     'convex_area', 
                     'area',
                     'convex_area',
                     'filled_area',
                     'eccentricity', 
                     'extent',
                     'eccentricity',
                     'major_axis_length',
                     'feret_diameter_max',
                     'inertia_tensor',
                     'perimeter',
                     'perimeter_crofton',
                     'minor_axis_length']
    
    return pd.DataFrame(regionprops_table(label_image, properties=feature_names))
```

```python
out = pd.DataFrame()

for index in glob('../images_masks_cropped/masks/*'):
    
    first = glob(index + '/*')[0]
    
    mask = imread(first)
    
    props = extract_props_from_image(mask)
    props['label'] = index.split('/')[-1]
    
    out = pd.concat([out, props], ignore_index=True)
```
