```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.preprocessing import MinMaxScaler

from glob import glob
```

```python
len(glob('dados/*'))
```

```python
mean_area = []
mean_equivalent_diameter = []
mean_minor_axis_length = []
mean_major_axis_length = []
mean_eccentricity = []
mean_convex_area = []
mean_equivalent_diameter = []
classes = []

for index in glob('dados/*'):
    
    df = pd.read_csv(index)
    target = index.split('/')[1].split('.')[0]
    
    mean_area.append(df.area.mean())
    mean_equivalent_diameter.append(df.equivalent_diameter.mean())
    mean_minor_axis_length.append(df.minor_axis_length.mean())
    mean_major_axis_length.append(df.major_axis_length.mean())
    mean_eccentricity.append(df.eccentricity.mean())
    mean_convex_area.append(df.convex_area.mean())
    classes.append(int(target))
    
```

```python
out = pd.DataFrame({
        'mean_area': mean_area,
        'mean_equivalent_diameter': mean_equivalent_diameter,
        'mean_minor_axis_length': mean_minor_axis_length,
        'mean_major_axis_length': mean_major_axis_length,
        'mean_eccentricity': mean_eccentricity,
        'mean_convex_area': mean_convex_area,
        'anilha': classes,
    })

out['target'] = 0
```

```python
broilers_weights = pd.read_csv('broilers_weights.csv')
```

```python
for index_anilha in broilers_weights.galinha:
    try:
        index = broilers_weights[broilers_weights.galinha == index_anilha].index[0]
        value = broilers_weights.loc[index].peso

        index_out = out[out.anilha == index_anilha].index[0]
        out['target'].iloc[index_out] = value
    except: pass
```
