```python
import pandas as pd
```

```python
out = pd.read_csv('dataset.csv')
broilers_weights = pd.read_csv('galinhas-pesos.csv')
```

```python
out['target'] = -1
```

```python
out.head()
```

```python
for anilha, peso in zip(broilers_weights.anilha,  broilers_weights.peso):

    try: 
        index = out[out.label == anilha].index[0]
        out.target.loc[index] = peso
    except: continue
```

```python
out.to_csv('features_means_regression.csv', index=None)
```
